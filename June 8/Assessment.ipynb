```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col,when

spark=SparkSession.builder\
.appName("DataFrame Exercises")\
.getOrCreate()
```

```python
#46
df_category=df.withColumn(
    "total_amount",
    col("quantity")*col("price")
).withColumn(
    "order_value_category",
    when(col("total_amount")>=50000,"High")
    .when(col("total_amount")>=20000,"Medium")
    .otherwise("Low")
)

df_category.show()
```

```python
#47
df_category.groupBy(
    "order_value_category"
).count().show()
```

```python
#48
df_final=df.withColumn(
    "total_amount",
    col("quantity")*col("price")
).withColumn(
    "discount",
    col("total_amount")*0.10
).withColumn(
    "final_amount",
    col("total_amount")-
    col("discount")
)
df_final.filter(
    df_final.order_status=="Delivered"
).select(
    "order_id",
    "customer_name",
    "final_amount"
).show()
```

```python
#49
df_final=df.withColumn(
    "total_amount",
    col("quantity")*col("price")
).withColumn(
    "discount",
    col("total_amount")*0.10
).withColumn(
    "final_amount",
    col("total_amount")-
    col("discount")
)

df_final.filter(
    df_final.order_status=="Delivered"
).groupBy(
    "city"
).sum("final_amount").show()
```

```python
#50
df_report=df.withColumn(
    "total_amount",
    col("quantity")*col("price")
).withColumn(
    "discount",
    col("total_amount")*0.10
).withColumn(
    "final_amount",
    col("total_amount")-
    col("discount")
)

df_report.select(
    "order_id",
    "customer_name",
    "city",
    "category",
    "product",
    "quantity",
    "price",
    "total_amount",
    "discount",
    "final_amount",
    "order_status"
).show()
```

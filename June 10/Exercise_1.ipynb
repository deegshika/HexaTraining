```python
df=Spark.read.csv(
    "data.csv",
    header=True,
    inferSchema=True

)
df.show()
```

```python
#11
df.filter(
df.status=="Confirmed"
).groupBy().sum(
"ticket_price"
).show()
```

```python
#12
df.groupBy(
"airline"
).avg(
"ticket_price"
).show()
```

```python
#13
df.groupBy().max(
"ticket_price"
).show()
```

```python
#14
df.groupBy().min(
"ticket_price"
).show()
```

```python
#15
df.orderBy(
df.ticket_price.desc()
).show()
```

```python
#16
df_tax=df.withColumn(
"tax",
df.ticket_price*0.05
)

df_tax.show()
```

```python
#17
df_final=df.withColumn(
"tax",
df.ticket_price*0.05
).withColumn(
"final_price",
df.ticket_price+
(df.ticket_price*0.05)
)

df_final.show()
```

```python
#18
from pyspark.sql.functions import when

df_category=df.withColumn(
"price_category",
when(
df.ticket_price>=30000,
"Premium"
).when(
df.ticket_price>=10000,
"Standard"
).otherwise(
"Budget"
)
)

df_category.show()
```

```python
#19
df_category.groupBy(
"price_category"
).count().show()
```

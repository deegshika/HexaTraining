```python
#18
from pyspark.sql.functions import when

df_status=df.withColumn(
"data_quality_status",
when(
col("customer_name").isNull()|
col("service_type").isNull()|
col("booking_amount").isNull(),
"Incomplete"
).otherwise(
"Complete"
)
)

df_status.show()
```

```python
#19
df_status.groupBy(
"data_quality_status"
).count().show()
```

```python
#20
df_status.filter(
df_status.data_quality_status==
"Complete"
).show()
```

```python
#21
df_status.filter(
df_status.data_quality_status==
"Incomplete"
).show()
```

```python
#22
df_tax=df.fillna({
"booking_amount":0
}).withColumn(
"tax",
col("booking_amount")*0.05
)

df_tax.show()
```

```python
#23
df_final=df.fillna({
"booking_amount":0
}).withColumn(
"tax",
col("booking_amount")*0.05
).withColumn(
"final_amount",
col("booking_amount")+
(col("booking_amount")*0.05)
)

df_final.show()
```

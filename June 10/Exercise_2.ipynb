```python
from pyspark.sql import SparkSession
Spark=SparkSession.builder\
.appName("Working with files")\
.getOrCreate()
```

```python
from google.colab import files
uploaded=files.upload()
```

```python
df=Spark.read.option(
"multiline",
"true"
).json("hotels.json")

df.show(truncate=False)

df.printSchema()
```

```python
#1
df.show()
```

```python
#2
df.select(
"hotel_name",
"city",
"rating"
).show()
```

```python
#3
df.filter(
df.category=="Luxury"
).show()
```

```python
#4
df.filter(
df.rating>4.5
).show()
```

```python
#5
df.filter(
df.rooms_available>15
).show()
```

```python
#6
df.filter(
df.price_per_night>10000
).show()
```

```python
#7
df.filter(
df.city.isin(
"Dubai",
"London"
)
).show()
```

```python
#8
df.filter(
df.contact.phone.isNull()
).show()
```

```python
#9
df.filter(
df.contact.email.isNull()
).show()
```

```python
#10
df.select(
"hotel_name",
"contact.phone",
"contact.email"
).show()
```

```python
from pyspark.sql import functions as F

#11
df.filter(
F.array_contains(df.amenities, "wifi")
).show()
```

```python
#12
from pyspark.sql import functions as F
df.filter(
F.array_contains(df.amenities,
"spa"
)
).show()
```

```python
#13
df.groupBy(
"city"
).count().show()
```

```python
#14
df.groupBy(
"category"
).count().show()
```

```python
#15
df.groupBy(
"category"
).avg(
"rating"
).show()
```

```python
#16
df.groupBy(
"city"
).avg(
"price_per_night"
).show()
```

```python
#17
df.groupBy().max(
"price_per_night"
).show()
```

```python
#18
df.withColumn(
"total_potential_revenue",
df.rooms_available*
df.price_per_night
).show()
```

```python
#19
df.orderBy(
df.rating.desc()
).show()
```

```python
#20
df_flat=df.select(
"hotel_id",
"hotel_name",
"city",
"category",
"rating",
"rooms_available",
"price_per_night",
"amenities",
"contact.phone",
"contact.email"
)

df_flat.write.mode(
"overwrite"
).parquet(
"hotels_flattened.parquet"
)
```

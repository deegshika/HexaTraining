```python
#61
from pyspark.sql.window import Window
from pyspark.sql.functions import col,sum,rank

window_spec=Window.orderBy(
    col("total_fee").desc()
)

fees_df=appointments_df.groupBy(
    "patient_id"
).sum(
    "consultation_fee"
).withColumnRenamed(
    "sum(consultation_fee)",
    "total_fee"
)

fees_df.withColumn(
    "rank",
    rank().over(window_spec)
).show()
```

```python
#62
from pyspark.sql.functions import dense_rank
fees_df.withColumn(
    "dense_rank",
    dense_rank().over(window_spec)
).show()
```

```python
#63
from pyspark.sql.functions import row_number
fees_df.withColumn(
    "row_number",
    row_number().over(window_spec)
).show()
```

```python
#91
patients_df=Spark.read.csv(
    "patients.csv",
    header=True,
    inferSchema=True
)

appointments_df=Spark.read.csv(
    "appointments.csv",
    header=True,
    inferSchema=True
)
```

```python
#92
preferences_df=Spark.read.option(
    "multiline",
    "true"
).json(
    "patient_preferences.json"
)
```

```python
#93
patients_clean=patients_df.fillna({
    "city":"Unknown",
    "blood_group":
    "Not Available"
})

appointments_clean=appointments_df.fillna({
    "consultation_fee":0
})

preferences_clean=preferences_df.fillna({
    "preferred_hospital":
    "Not Provided"
})
```

```python
#94
final_df=patients_clean.join(
    appointments_clean,
    "patient_id",
    "left"
).join(
    preferences_clean,
    "patient_id",
    "left"
)

final_df.show()
```

```python
#95
final_df=final_df.withColumn(
    "age_group",
    when(
        col("age")<30,
        "Young"
    ).when(
        col("age")<=50,
        "Adult"
    ).otherwise(
        "Senior"
    )
)

final_df.show()
```

```python
department_revenue = appointments_clean.groupBy(
    "department"
).sum(
    "consultation_fee"
)

department_revenue.show()
```

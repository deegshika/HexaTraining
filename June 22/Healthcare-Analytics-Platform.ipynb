```python
# #1 File paths
patients_path="/Volumes/ey_databricks/default/hospital/patients.csv"
doctors_path="/Volumes/ey_databricks/default/hospital/doctors.csv"
appointments_path="/Volumes/ey_databricks/default/hospital/appoinments.csv"
preferences_path="/Volumes/ey_databricks/default/hospital/patient_preferences.json"
```

```python
# #2 Read patients.csv
patients_df=spark.read.format(
    "csv"
).option(
    "header",
    "true"
).option(
    "inferSchema",
    "true"
).load(
    patients_path
)

display(patients_df)
```

```python
# #3 Read doctors.csv
doctors_df=spark.read.format(
    "csv"
).option(
    "header",
    "true"
).option(
    "inferSchema",
    "true"
).load(
    doctors_path
)

display(doctors_df)
```

```python
# #4 Read appointments.csv
appointments_df=spark.read.format(
    "csv"
).option(
    "header",
    "true"
).option(
    "inferSchema",
    "true"
).load(
    appointments_path
)

display(appointments_df)
```

```python
# #5 Read patient_preferences.json
preferences_df=spark.read.format(
    "json"
).option(
    "multiline",
    "true"
).load(
    preferences_path
)

display(preferences_df)
```

```python
# #6 Display schema of all datasets
patients_df.printSchema()
doctors_df.printSchema()
appointments_df.printSchema()
preferences_df.printSchema()
```

```python
# #7 Save raw data as Bronze Delta tables
patients_df.write.format(
    "delta"
).mode(
    "overwrite"
).saveAsTable(
    "bronze_patients"
)

doctors_df.write.format(
    "delta"
).mode(
    "overwrite"
).saveAsTable(
    "bronze_doctors"
)

appointments_df.write.format(
    "delta"
).mode(
    "overwrite"
).saveAsTable(
    "bronze_appointments"
)

preferences_df.write.format(
    "delta"
).mode(
    "overwrite"
).saveAsTable(
    "bronze_patient_preferences"
)
```

```python
# Verify Bronze tables
display(
    spark.table(
        "bronze_patients"
    )
)

display(
    spark.table(
        "bronze_doctors"
    )
)

display(
    spark.table(
        "bronze_appointments"
    )
)

display(
    spark.table(
        "bronze_patient_preferences"
    )
)
```

```python
# #8 Handle missing values if present
from pyspark.sql.functions import col,when,month,to_date

patients_clean_df=patients_df.fillna(
    {
        "city":"Unknown",
        "state":"Unknown",
        "insurance_status":"Unknown"
    }
)

doctors_clean_df=doctors_df.fillna(
    {
        "department":"Unknown",
        "city":"Unknown",
        "consultation_fee":0
    }
)

appointments_clean_df=appointments_df.fillna(
    {
        "diagnosis":"Unknown",
        "bill_amount":0,
        "status":"Unknown"
    }
)

display(patients_clean_df)
display(doctors_clean_df)
display(appointments_clean_df)
```

```python
# #9 Flatten patient_preferences.json
preferences_flat_df=preferences_df.select(
    col("patient_id"),
    col("preferred_hospital"),
    col("contact.phone").alias("phone"),
    col("contact.email").alias("email")
)

display(preferences_flat_df)
```

```python
# #10 Join patients with preferences
patients_preferences_df=patients_clean_df.join(
    preferences_flat_df,
    on="patient_id",
    how="left"
)

display(patients_preferences_df)
```

```python
# #11 Join appointments with patients
appointments_patients_df=appointments_clean_df.join(
    patients_clean_df,
    on="patient_id",
    how="left"
)

display(appointments_patients_df)
```

```python
# #12 Join appointments with doctors
hospital_joined_df=appointments_patients_df.join(
    doctors_clean_df,
    on="doctor_id",
    how="left"
)

display(hospital_joined_df)
```

```python
# #13 Create final_bill = bill_amount + consultation_fee
hospital_joined_df=hospital_joined_df.withColumn(
    "final_bill",
    col("bill_amount")+col("consultation_fee")
)

display(hospital_joined_df)
```

```python
# #14 Create appointment_month from appointment_date
hospital_joined_df=hospital_joined_df.withColumn(
    "appointment_date",
    to_date(col("appointment_date"))
).withColumn(
    "appointment_month",
    month(col("appointment_date"))
)

display(hospital_joined_df)
```

```python
# #15 Create patient_age_group
hospital_joined_df=hospital_joined_df.withColumn(
    "patient_age_group",
    when(col("age")>=50,"Senior")
    .when(col("age")>=30,"Adult")
    .otherwise("Young")
)

display(hospital_joined_df)
```

```python
# #16 Save transformed data as Silver Delta tables
patients_clean_df.write.format(
    "delta"
).mode(
    "overwrite"
).saveAsTable(
    "silver_patients"
)

doctors_clean_df.write.format(
    "delta"
).mode(
    "overwrite"
).saveAsTable(
    "silver_doctors"
)

appointments_clean_df.write.format(
    "delta"
).mode(
    "overwrite"
).saveAsTable(
    "silver_appointments"
)

preferences_flat_df.write.format(
    "delta"
).mode(
    "overwrite"
).saveAsTable(
    "silver_patient_preferences"
)

_cols = hospital_joined_df.columns
_seen = {}
_new_cols = []
for _c in _cols:
    if _c in _seen:
        _seen[_c] += 1
        _new_cols.append(f"doctor_{_c}")
    else:
        _seen[_c] = 0
        _new_cols.append(_c)
hospital_joined_df.toDF(*_new_cols).write.format(
    "delta"
).mode(
    "overwrite"
).saveAsTable(
    "silver_hospital_analytics"
)
```

```python
# Verify Silver tables
display(spark.table("silver_patients"))
display(spark.table("silver_doctors"))
display(spark.table("silver_appointments"))
display(spark.table("silver_patient_preferences"))
display(spark.table("silver_hospital_analytics"))
```

```python
# #23 Rank doctors by revenue
from pyspark.sql.window import Window
from pyspark.sql.functions import sum,rank,row_number,col

doctor_revenue_df=hospital_joined_df.groupBy(
    "doctor_id",
    "doctor_name",
    "department"
).agg(
    sum(
        "final_bill"
    ).alias(
        "doctor_revenue"
    )
)

doctor_window=Window.orderBy(
    col(
        "doctor_revenue"
    ).desc()
)

doctor_rank_df=doctor_revenue_df.withColumn(
    "doctor_rank",
    rank().over(
        doctor_window
    )
)

display(doctor_rank_df)
```

```python
# #24 Rank departments by revenue
department_revenue_df=hospital_joined_df.groupBy(
    "department"
).agg(
    sum(
        "final_bill"
    ).alias(
        "department_revenue"
    )
)

department_window=Window.orderBy(
    col(
        "department_revenue"
    ).desc()
)

department_rank_df=department_revenue_df.withColumn(
    "department_rank",
    rank().over(
        department_window
    )
)

display(department_rank_df)
```

```python
# #25 Find top 3 patients by billing
patient_billing_df=hospital_joined_df.groupBy(
    "patient_id",
    "patient_name"
).agg(
    sum(
        "final_bill"
    ).alias(
        "total_billing"
    )
)

patient_window=Window.orderBy(
    col(
        "total_billing"
    ).desc()
)

top3_patients_df=patient_billing_df.withColumn(
    "patient_rank",
    rank().over(
        patient_window
    )
).filter(
    col(
        "patient_rank"
    )<=3
)

display(top3_patients_df)
```

```python
# #26 Find top doctor in each department
doctor_dept_revenue_df=hospital_joined_df.groupBy(
    "department",
    "doctor_id",
    "doctor_name"
).agg(
    sum(
        "final_bill"
    ).alias(
        "doctor_revenue"
    )
)

dept_window=Window.partitionBy(
    "department"
).orderBy(
    col(
        "doctor_revenue"
    ).desc()
)

top_doctor_department_df=doctor_dept_revenue_df.withColumn(
    "dept_rank",
    row_number().over(
        dept_window
    )
).filter(
    col(
        "dept_rank"
    )==1
)

display(top_doctor_department_df)
```

```python
# #27 Create running revenue by appointment_date
daily_revenue_df=hospital_joined_df.groupBy(
    "appointment_date"
).agg(
    sum(
        "final_bill"
    ).alias(
        "daily_revenue"
    )
)

running_window=Window.orderBy(
    "appointment_date"
).rowsBetween(
    Window.unboundedPreceding,
    Window.currentRow
)

running_revenue_df=daily_revenue_df.withColumn(
    "running_revenue",
    sum(
        "daily_revenue"
    ).over(
        running_window
    )
)

display(running_revenue_df)
```

```python
# Fix duplicate columns before Delta write
from pyspark.sql.functions import col

hospital_delta_df=appointments_clean_df.alias("a").join(
    patients_clean_df.alias("p"),
    col("a.patient_id")==col("p.patient_id"),
    "left"
).join(
    doctors_clean_df.alias("d"),
    col("a.doctor_id")==col("d.doctor_id"),
    "left"
).select(
    col("a.appointment_id"),
    col("a.patient_id"),
    col("a.doctor_id"),
    col("a.appointment_date"),
    col("a.diagnosis"),
    col("a.bill_amount"),
    col("a.status"),
    col("p.patient_name"),
    col("p.city").alias("patient_city"),
    col("p.state"),
    col("p.age"),
    col("p.gender"),
    col("p.insurance_status"),
    col("d.doctor_name"),
    col("d.department"),
    col("d.city").alias("doctor_city"),
    col("d.consultation_fee")
).withColumn(
    "final_bill",
    col("bill_amount")+col("consultation_fee")
)

display(hospital_delta_df)
```

```python
# #33 Perform SCD Type 1 merge using updated patient city/insurance data
from delta.tables import DeltaTable

updated_patients_data=[
    ("P101","Rahul Sharma","Secunderabad","Telangana",35,"Male","Inactive"),
    ("P104","Sneha Patel","Noida","Delhi",31,"Female","Active")
]

updated_patients_columns=[
    "patient_id",
    "patient_name",
    "city",
    "state",
    "age",
    "gender",
    "insurance_status"
]

updated_patients_df=spark.createDataFrame(
    updated_patients_data,
    updated_patients_columns
)

delta_patients=DeltaTable.forName(
    spark,
    "silver_patients"
)

delta_patients.alias(
    "target"
).merge(
    updated_patients_df.alias("source"),
    "target.patient_id=source.patient_id"
).whenMatchedUpdate(
    set={
        "patient_name":"source.patient_name",
        "city":"source.city",
        "state":"source.state",
        "age":"source.age",
        "gender":"source.gender",
        "insurance_status":"source.insurance_status"
    }
).whenNotMatchedInsertAll().execute()
```

```python
#76
revenue_df=df.withColumn(
    "revenue",
    col("quantity")*
    col("price")
)

window_spec=Window.orderBy(
    col("revenue").desc()
)

revenue_df.withColumn(
    "row_number",
    row_number().over(
        window_spec
    )
).show()
```

```python
#77
revenue_df.withColumn(
    "rank",
    rank().over(
        window_spec
    )
).show()
```

```python
#81
city_window=Window.partitionBy(
    "city"
).orderBy(
    col("revenue").desc()
)

revenue_df.withColumn(
    "rank",
    row_number().over(
        city_window
    )
).filter(
    col("rank")==1
).show()
```

```python
#82
category_window=Window.partitionBy(
    "category"
).orderBy(
    col("revenue").desc()
)

revenue_df.withColumn(
    "rank",
    row_number().over(
        category_window
    )
).filter(
    col("rank")==1
).show()
```

```python
#83
running_window=Window.orderBy(
    "order_id"
)

revenue_df.withColumn(
    "running_total",
    sum(
        "revenue"
    ).over(
        running_window
    )
).show()
```

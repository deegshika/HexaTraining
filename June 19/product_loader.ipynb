```python
products = [
    (101, "Laptop", 55000),
    (102, "Headphones", 2500),
    (103, "Smart Watch", 6000),
    (104, "Mouse", 1000)
]

columns = [
    "product_id",
    "product_name",
    "price"
]

df = spark.createDataFrame(products, columns)

display(df)
```

```python
df.createOrReplaceTempView("products")

print("Temp View Created")
```

```python
%sql
SELECT
AVG(price) AS average_price
FROM products
```

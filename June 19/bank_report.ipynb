```python
accounts = [
    (1, "Chennai", 10000),
    (2, "Chennai", 15000),
    (3, "Bangalore", 12000),
    (4, "Coimbatore", 8000),
    (5, "Bangalore", 18000)
]

columns = [
    "account_id",
    "city",
    "balance"
]

df = spark.createDataFrame(accounts, columns)

display(df)

report = df.groupBy("city").sum("balance")

display(report)

print("Banking Report Generated")
```

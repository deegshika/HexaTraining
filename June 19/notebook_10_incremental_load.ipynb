```python
day1_data = [

(1,"Rahul",8500),
(2,"Priya",22000),
(3,"Amit",9000),
(4,"Sneha",15000),
(5,"Farhan",7500)

]

columns = [
    "booking_id",
    "customer_name",
    "ticket_price"
]

day1_df = spark.createDataFrame(
    day1_data,
    columns
)

display(day1_df)
```

```python
day1_df.write \
    .format("delta") \
    .mode("overwrite") \
    .saveAsTable("flight_bookings")
```

```python
day2_data = [

(2,"Priya",25000),      # updated

(4,"Sneha",17000),      # updated

(6,"Arjun",10000),      # new

(7,"Meera",17000),      # new

(8,"Kiran",9500),       # new

(9,"Nisha",28000),      # new

(10,"David",8000)       # new

]

day2_df = spark.createDataFrame(
    day2_data,
    columns
)

display(day2_df)
```

```python
day2_df.createOrReplaceTempView(
    "booking_updates"
)
```

```python
spark.sql("""

MERGE INTO flight_bookings target

USING booking_updates source

ON target.booking_id = source.booking_id

WHEN MATCHED THEN

UPDATE SET
target.customer_name = source.customer_name,
target.ticket_price = source.ticket_price

WHEN NOT MATCHED THEN

INSERT *

""")
```

```python
spark.sql("""

SELECT *
FROM flight_bookings
ORDER BY booking_id

""").show()
```

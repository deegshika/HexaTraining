```python
from pyspark.sql.functions import col

df_bookings = df_bookings.withColumn(
    "revenue",
    col("ticket_price")
)

display(df_bookings)
```

```python
from pyspark.sql.functions import when

df_bookings = df_bookings.withColumn(
    "price_band",
    when(col("ticket_price") > 20000, "Premium")
    .when(col("ticket_price") > 10000, "Standard")
    .otherwise("Budget")
)

display(df_bookings)
```
